```python
def combine_indicators(close, window=14, window_bb = 20, entry=30, exit=70):
    trend = np.full(close.shape, np.nan)    
    rsi = vbt.RSI.run(close, window=window, short_name="rsi").rsi
#     rsi_above = rsi.rsi_above(exit).to_numpy()
#     rsi_below = rsi.rsi_below(entry).to_numpy()
    bb = vbt.BBANDS.run(close, window = window_bb)
    
    upper_band = bb.upper
    lower_band = bb.lower
 
    
    for x in range(len(close)):
        if rsi.values[x] > exit and  upper_band.values[x] < close[x]:  # sell 1
            trend[x] = 1
        elif rsi.values[x] < entry and lower_band.values[x] > close[x]: # buy -1
            trend[x] = -1
        else:
            trend[x] = 0
            
    return trend
```

```python
myInd = vbt.IndicatorFactory(
    class_name="Combination",
    short_name="comb",
    input_names=["close"],
    param_names=["window", "window_bb","entry", "exit"],
    output_names=["ind"]
).from_apply_func(
    combine_indicators,
    window=14,
    window_bb = 20,
    entry=30,
    exit=70
)
 
 

res = myInd.run(
    btc_price_1d,
    window=np.arange(2,20,step=1,dtype=int),
    window_bb = np.arange(10,20,step=1,dtype=int),
    entry=np.arange(10,50,step=10,dtype=int),
    exit=np.arange(50,90,step=10,dtype=int),
    
    param_product=True,
)
 
```
